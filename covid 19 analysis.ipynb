```python
df.fillna(0,inplace=True)
```

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
```

```python

top_confirmed = df.sort_values(by='Confirmed', ascending=False).head(10)
top_confirmed[['Country/Region', 'Confirmed']]
```

```python
top_recovered = df.sort_values(by='Recovered', ascending=False).head(10)
top_recovered[['Country/Region', 'Recovered']]
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_confirmed['Country/Region'], top_confirmed['Confirmed'])
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.show()
```

```python
total_confirmed = df['Confirmed'].sum()
total_deaths = df['Deaths'].sum()
total_recovered = df['Recovered'].sum()
total_active = df['Active'].sum()

plt.figure(figsize=(6,6))
plt.pie(
    [total_active, total_recovered, total_deaths],
    labels=['Active', 'Recovered', 'Deaths'],
    autopct='%1.1f%%'
)
plt.title("Global COVID-19 Case Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['Confirmed','Deaths','Recovered','Active']].corr(),
    annot=True
)
plt.title("COVID-19 Correlation Heatmap")
plt.show()
```
